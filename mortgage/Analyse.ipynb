#### Summary
There are `500,000` instances of the features and labels. There are `22` columns in the features file and `2` columns in the labels file including the `row_id` where every instance has it's respective label `accepted` referenced in the training_labels file.

All the features except the `co_applicant` are numeric types with some being integers and others being floating points. The `co_applicant` is a boolean type indicating whether the application had a co-applicant. The label attribute `accepted` is an integer $\{ 0=not accepted,1=accepted \}$ indicating whether the loan application was accepted or not.
